```python
class Playlist():
    def __init__(self,song_list):
        self.song_list = song_list

    def afegir_canco(self, song):
        self.song_list.append(song)

    def durada_total(self, song_list):
        total= 0
        for song in song_list:
            total += song.durada
        return total

    def mostrar_playlist(self):
        for song in self.song_list:
            print(f"TÍTOL: {song.titol}, COMPOSITOR: {song.autor}, DURADA: {song.durada} segons")
    
    def canco_mes_llarga(self):
        canco_mes_llarga = self.song_list
        for song in self.song_list:
            if song.durada > canco_mes_llarga.durada:
                canco_mes_llarga = song
        return canco_mes_llarga
```

```python
class Personatge:
    def __init__(self, nom, vida, forca, nivell):
        self.nom = nom
        self.vida = vida
        self.forca = forca
        self.nivell = nivell
    def atacar(self, pers2):
        if not isinstance(pers2, Personatge):
            print("L'enemic no és un personatge")
        else:
            if pers2.vida <=0:
                print("Ha mort l'enemic")
            else:
                pers2.vida= pers2.vida - self.forca
    def pujar_nivell(self):
        self.nivell+=1
        self.forca+=2
        self.vida+=5
    def esta_viu(self):
        if self.vida <=0:
            return False
        else:
            return True
    def saber_info(self):
        return f"Personatge{self.nom}, Vida{self.vida}, Força{self.forca}, Nivell{self.nivell}"


el_bicho=Personatge(nom="bicharraco", vida=100, forca=10, nivell=1)
joan_pradells=Personatge(nom="jhonny", vida=100, forca=70, nivell= 17)




while (el_bicho.esta_viu() == True) and (joan_pradells.esta_viu() == True):
    el_bicho.atacar(joan_pradells)
    joan_pradells.atacar(el_bicho)
if el_bicho.esta_viu()==True:
    print("bicharraco ha guanyat")
else:
    print("jhonny ha guanyat")
```
